Sentiment Analysis

```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
data = ["I love you"]
sentiment_pipeline(data)
```

```python
Data_Frame.dtypes
```

```python
Data_Frame['content'] = Data_Frame['content'].astype('str')
```

```python
Data_Frame['result'] = Data_Frame['content'].apply(lambda x: sentiment_pipeline(x))
```

```python
Data_Frame.head()
```

```python
Data_Frame.shape
```

```python
Data_Frame['sentiment']= Data_Frame['result'].apply(lambda x: (x[0]['label']))
Data_Frame['sentiment_score']= Data_Frame['result'].apply(lambda x: (x[0]['score']))
```
